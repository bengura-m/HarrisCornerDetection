# Extracting frames from the internet

```python
import cv2

video_path = 'preview.mp4'
output_path = 'output_frames/'

cap = cv2.VideoCapture(video_path)
frame_count = 0
frame_interval = 2  # Extract a frame every 2 seconds
time_elapsed = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    time_elapsed += 1
    if time_elapsed < frame_interval * cap.get(cv2.CAP_PROP_FPS):
        continue
    
    frame_count += 1
    frame_filename = f'{output_path}frame_{frame_count:04d}.jpg'
    cv2.imwrite(frame_filename, frame)
    
    time_elapsed = 0  # Reset time_elapsed after capturing a frame
    
    if frame_count >= 5:
        break

cap.release()
cv2.destroyAllWindows()
```

# Sorting frames into a collage

```python
import cv2
import numpy as np
import os

#Function calculating the color variance
def calculate_color_variance(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    return np.var(v)

# Function to calculate the edge strength
def calculate_edge_strength(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    return np.sum(edges)

# Function to calculate the color histogram variance
def calculate_histogram_variance(image):
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    hist = cv2.normalize(hist, hist).flatten()
    return np.var(hist)

# Function to sort the frames by color and edge. (lower variance towards center) and edge strength (higher strength towards boundary)
def sort_frames(frames):
    frames.sort(key=lambda frame: (calculate_color_variance(frame), calculate_edge_strength(frame), calculate_histogram_variance(frame)))

# Main function
def main():
    output_path = 'output_frames/'
    frames = []
    
    # This loads frames from the output_path
    for filename in os.listdir(output_path):
        if filename.endswith('.jpg'):
            frame = cv2.imread(os.path.join(output_path, filename))
            frames.append(frame)
    
    # Sort frames based on color variance and edge strength
    sort_frames(frames)
    
    # Resize frames so they fit 
    target_sizes = [(400, 200)] * 3 + [(600, 300)] * 2
    resized_frames = [cv2.resize(frame, size) for frame, size in zip(frames, target_sizes)]
    
    # Creating horizontal stacks
    horizontal_stacks = [np.hstack(resized_frames[i:i + 3]) for i in range(0, len(resized_frames), 3)]
    
    # Creating vertical stack
    final_collage = np.vstack(horizontal_stacks)
    
    # Show the final collage
    cv2.imshow("Final Collage", final_collage)
    
    collage_filename = "final_collage.jpg"
    cv2.imwrite(collage_filename, final_collage)

    print(f"Collage saved as '{collage_filename}'")
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import os
import random
from skimage import io, color, transform
from scipy.ndimage import sobel

# Function calculating the color variance
def calculate_color_variance(image):
    hsv = color.rgb2hsv(image)
    v = hsv[:, :, 2]
    return np.var(v)

# Function to calculate the edge strength
def calculate_edge_strength(image):
    gray = color.rgb2gray(image)
    edges = sobel(gray)
    return np.sum(edges)

# Function to calculate the color histogram variance
def calculate_histogram_variance(image):
    hist, _ = np.histogram(image.ravel(), bins=8, range=[0, 256])
    hist = hist / hist.sum()
    return np.var(hist)

# Function to sort the frames by color and edge. (lower variance towards center) and edge strength (higher strength towards boundary)
def sort_frames(frames):
    frames.sort(key=lambda frame: (calculate_color_variance(frame), calculate_edge_strength(frame), calculate_histogram_variance(frame)))

# Main function
def main():
    output_path = 'output_frames/'
    frames = []
    
    # This loads frames from the output_path
    for filename in os.listdir(output_path):
        if filename.endswith('.jpg'):
            frame = io.imread(os.path.join(output_path, filename))
            frames.append(frame)
    
    # Sort frames based on color variance and edge strength
    sort_frames(frames)
    
    # Resize frames so they fit 
    target_sizes = [(200, 100)] * 3 + [(300, 150)] * 2
    resized_frames = [transform.resize(frame, size) for frame, size in zip(frames, target_sizes)]
    
    # Creating horizontal stacks
    horizontal_stacks = [np.hstack(resized_frames[i:i + 3]) for i in range(0, len(resized_frames), 3)]
    
    # Creating vertical stack
    final_collage = np.vstack(horizontal_stacks)
    
       # Convert the image data to unsigned 8-bit integers in the range [0, 255]
    final_collage = (final_collage * 255).astype(np.uint8)
    
    # Display or save the final collage
    io.imshow(final_collage)
    collage_filename = "final_collage.jpg"
    io.imsave(collage_filename, final_collage)
    print(f"Collage saved as '{collage_filename}'")

if __name__ == "__main__":
    main()
```
