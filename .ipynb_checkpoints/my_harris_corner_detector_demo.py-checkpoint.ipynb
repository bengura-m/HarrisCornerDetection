```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import data,color  # To input standard images
from matplotlib import pyplot as plt  # To plot images 
from skimage.color import rgb2gray  # To convert RGB image to grayscale
from skimage.feature import corner_harris, corner_subpix, corner_peaks  # Library for Harris corner detection
```

# My Harris Corner Detector

```python
#Loading the images from skimage
astronaut_image = data.astronaut()
checkerboard_image = data.checkerboard()
```

```python
# Function for my harris corner detector
def my_harris_corner_detector_demo(input_image, window_size, k, threshold):
    
     # Convert to grayscale if not already grayscale
        
    if len(input_image.shape) == 3:  # Check if the image has color channels
        gray = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
    else:
        gray = input_image  # The input image is already grayscale

    # applying gaussian blur to the grayscale image   
    img_gaussian = cv2.GaussianBlur(gray, (3, 3), 0)
        
    # extractong image dimensions
    height = input_image.shape[0]  # Height
    width = input_image.shape[1]   # Width
    
    # Initialize the response matrix
    matrix_R = np.zeros((height, width))
    
    # Step 1 - Calculate the x and y image derivatives (dx and dy)
    dx = cv2.Sobel(img_gaussian, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(img_gaussian, cv2.CV_64F, 0, 1, ksize=3)
    
    # Step 2 - Calculate product and second derivatives (dx2, dy2, and dxy)
    dx2 = np.square(dx)
    dy2 = np.square(dy)
    dxy = dx * dy

    offset = int(window_size / 2)
    
    # Step 3 - Calculate the sum of products of derivatives for each pixel (Sx2, Sy2, and Sxy)
    print("Finding Corners...")
    for y in range(offset, height - offset):
        for x in range(offset, width - offset):
            Sx2 = np.sum(dx2[y - offset:y + 1 + offset, x - offset:x + 1 + offset])
            Sy2 = np.sum(dy2[y - offset:y + 1 + offset, x - offset:x + 1 + offset])
            Sxy = np.sum(dxy[y - offset:y + 1 + offset, x - offset:x + 1 + offset])

            # Step 4 - Define the matrix H(x,y)=[[Sx2,Sxy],[Sxy,Sy2]]
            H = np.array([[Sx2, Sxy], [Sxy, Sy2]])

            # Step 5 - Calculate the response function (R=det(H)-k(Trace(H))^2)
            det = np.linalg.det(H)
            tr = np.matrix.trace(H)
            R = det - k * (tr ** 2)
            matrix_R[y - offset, x - offset] = R
    
    # Step 6 - Apply a threshold and draw corners on the image
    cv2.normalize(matrix_R, matrix_R, 0, 1, cv2.NORM_MINMAX)
    for y in range(offset, height - offset):
        for x in range(offset, width - offset):
            value = matrix_R[y, x]
            if value > threshold:
                cv2.circle(input_image, (x, y), 3, (0, 255, 0))

    # Display the result
    plt.figure("My Harris detector")
    plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    plt.title("My Harris detector")
    plt.xticks([]), plt.yticks([])
#     plt.savefig('My_harris_detector-thresh_%s.png' % threshold, bbox_inches='tight')
    plt.show()
    
    return input_image
```

```python
# Display the astronaut image
plt.imshow(astronaut_image)
plt.title('Astronaut Image')
plt.axis('off')  # Turn off axes
plt.show()

# Display the checkerboard image
plt.imshow(checkerboard_image, cmap='gray')  # Use grayscale colormap
plt.title('Checkerboard Image')
plt.axis('off')  # Turn off axes
plt.show()
```

```python
my_astronaut_corners = my_harris_corner_detector_demo(astronaut_image, 7, 0.04, 0.50) 
my_checkerboard_corners = my_harris_corner_detector_demo(checkerboard_image,6,0.04,0.50)
```

```python
# Apply corner for comparison
# TODO: get grayscale image of input image
gray_astronaut_image = color.rgb2gray(astronaut_image)
astronaut_harris_response = corner_harris(gray_astronaut_image)
astronaut_peaks = corner_peaks(astronaut_harris_response, min_distance=5)

checkerboard_harris_response = corner_harris(checkerboard_image)
checkerboard_peaks = corner_peaks(checkerboard_harris_response, min_distance=5)
```

```python
# Display the detected corners
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(astronaut_image, cv2.COLOR_BGR2RGB))
plt.title("Astronaut Image")
plt.scatter(my_astronaut_corners[:, 0], my_astronaut_corners[:, 1], c='red', marker='o')
plt.scatter(astronaut_peaks[:, 1], astronaut_peaks[:, 0], c='blue', marker='+')

plt.subplot(2, 2, 2)
plt.imshow(checkerboard_image, cmap='gray')
plt.title("Checkerboard Image")
plt.scatter(my_checkerboard_corners[:, 0], my_checkerboard_corners[:, 1], c='red', marker='o')
plt.scatter(checkerboard_peaks[:, 1], checkerboard_peaks[:, 0], c='blue', marker='+')

plt.tight_layout()
plt.show()
```
